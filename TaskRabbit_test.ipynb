```r
#1. How many recommendation sets are in this data sample?

by_recommendationId <-  group_by(data, recommendation_id)
summary_recommendationId <- summarise(by_recommendationId, count=n())
count(summary_recommendationId)
```

```r
#3. How many total unique Taskers are there in this data sample?

by_taskerId <-  group_by(data, tasker_id)
summary_taskerId <- arrange(summarise(by_taskerId, count=n()), tasker_id)
count(summary_taskerId)
```

```r
#4. Which Tasker has been shown the most?
most_shown <- with(summary_taskerId, tasker_id[count==max(count)])
most_shown
```

```r
#4. Which Tasker has been shown the least?
least_shown <- with(summary_taskerId, tasker_id[count==min(count)])
least_shown
```

```r
#5. Which Tasker has been hired the most?
hired_taskerId <- arrange(summarise(by_taskerId, sum = sum(hired)),tasker_id)
highest_hired <- with(hired_taskerId, tasker_id[sum==max(sum)])
highest_hired
```

```r
#5. Which Tasker has been hired the least?
least_hired <- with(hired_taskerId, tasker_id[sum==min(sum)])
least_hired
```

```r
#Total no.of Taskers who were least hired
length(least_hired)
```

```r
#6. If we define the "Tasker conversion rate" as the number of times a Tasker has been hired, 
#out of the number of times the Tasker has been shown, how many Taskers have a conversion rate of 100%

conversion_rate<-mutate(hired_taskerId, conversion_rate = (hired_taskerId$sum)/(summary_taskerId$count))
conversion_rate
```

```r
# Total no.of Taskers who has 100% Conversion rate
conversion_rate_100 <- with(conversion_rate,tasker_id[conversion_rate==1])
length(conversion_rate_100)
```

```r
#7. Would it be possible for all Taskers to have a conversion rate of 100%. Please explain your reasoning.

# No.
# Reason-1: In the Data Description provided, it is stated that "Only 1 tasker out of a set of recommendations can be hired".
#           It clearly shows that not every tasker(only one tasker) in the respective recommendation_id group can be hired. 
#           Then, there is no possibility of all taskers to show and then hired all the times.
#           Hence, the conversion rate cannot be 100% for all Taskers.
# Reason-2: Out of the 830 Unique tasker_ids, only 312 taskers were hired atleast once when they were shown. 
#           Remaining 518 taskers were shown multiple times, but not hired at all. 
#           This again shows that there is no possibility of having 100% conversion rate for all the Taskers.
```

```r
#8. For each category, what is the average position of the Tasker who is hired?

filtered_hired <- filter(data, hired==1) # Filtering only who were hired
filtered_hired_arr <-arrange(filtered_hired,category,tasker_id) # Sorting w.r.t tasker_id
filtered_hired_arr
```

```r
#Finding the Average Position of each Tasker who is hired
Average_position <- group_by(filtered_hired_arr,category,tasker_id)%>% summarize(average_position = mean(position))
Average_position
```
